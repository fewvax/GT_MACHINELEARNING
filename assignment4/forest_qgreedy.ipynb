```python
import numpy as np
import pandas as pd
import gym
from gym import wrappers
import time
import sys
import matplotlib.pyplot as plt
from hiive.mdptoolbox import mdp, example
```

```python
def Forest_Experiments_ep(size):
    print('Q LEARNING WITH FOREST MANAGEMENT')
    P, R = example.forest(S=size, r1=10, r2=50, p = 0.1)
    value_f = []
    policy = []
    iters = []
    time_array = []
    Q_table = []
    rew_array = []
    for epsilon in [0.1,0.3,0.5,0.7,0.9]:
        st = time.time()
        pi = mdp.QLearning(P,R,0.1, epsilon=epsilon)
        end = time.time()
        temp_df = pd.DataFrame(pi.run())
        rew_array.append(temp_df['Mean V'].to_numpy())
        value_f.append(np.mean(pi.V))
        policy.append(pi.policy)
        time_array.append(end-st)
        Q_table.append(pi.Q)

    plt.plot(range(0,10000), rew_array[0],label='epsilon=0.1')
    plt.plot(range(0,10000), rew_array[1],label='epsilon=0.3')
    plt.plot(range(0,10000), rew_array[2],label='epsilon=0.5')
    plt.plot(range(0,10000), rew_array[3],label='epsilon=0.7')
    plt.plot(range(0,10000), rew_array[4],label='epsilon=0.9')
    plt.legend()
    plt.xlabel('Iterations')
    plt.grid()
    plt.title('Forest Management - Q Learning - Decaying Epsilon')
    plt.ylabel('Average Reward')
    plt.show()

    plt.subplot(1,6,1)
    plt.imshow(Q_table[0][:20,:])
    plt.title('0.1')

    plt.subplot(1,6,2)
    plt.title('0.3')
    plt.imshow(Q_table[1][:20,:])

    plt.subplot(1,6,3)
    plt.title('0.5')
    plt.imshow(Q_table[2][:20,:])

    plt.subplot(1,6,4)
    plt.title('0.7')
    plt.imshow(Q_table[3][:20,:])

    plt.subplot(1,6,5)
    plt.title('0.9')
    plt.imshow(Q_table[4][:20,:])
   

    return



```

```python
def Forest_Experiments_al(size):
    print('Q LEARNING WITH FOREST MANAGEMENT')
    P, R = example.forest(S=size, r1=10, r2=50, p = 0.1)
    value_f = []
    policy = []
    iters = []
    time_array = []
    Q_table = []
    rew_array = []
    for a in [0.1,0.3,0.5,0.7,0.9]:
        st = time.time()
        pi = mdp.QLearning(P,R,0.1, alpha=a)
        end = time.time()
        temp_df = pd.DataFrame(pi.run())
        rew_array.append(temp_df['Mean V'].to_numpy())
        value_f.append(np.mean(pi.V))
        policy.append(pi.policy)
        time_array.append(end-st)
        Q_table.append(pi.Q)

    plt.plot(range(0,10000), rew_array[0],label='alpha=0.1')
    plt.plot(range(0,10000), rew_array[1],label='alpha=0.3')
    plt.plot(range(0,10000), rew_array[2],label='alpha=0.5')
    plt.plot(range(0,10000), rew_array[3],label='alpha=0.7')
    plt.plot(range(0,10000), rew_array[4],label='alpha=0.9')
    plt.legend()
    plt.xlabel('Iterations')
    plt.grid()
    plt.title('Forest Management - Q Learning - Alpha')
    plt.ylabel('Average Reward')
    plt.show()

    plt.subplot(1,6,1)
    plt.imshow(Q_table[0][:20,:])
    plt.title('0.1')

    plt.subplot(1,6,2)
    plt.title('0.3')
    plt.imshow(Q_table[1][:20,:])

    plt.subplot(1,6,3)
    plt.title('0.5')
    plt.imshow(Q_table[2][:20,:])

    plt.subplot(1,6,4)
    plt.title('0.7')
    plt.imshow(Q_table[3][:20,:])

    plt.subplot(1,6,5)
    plt.title('0.9')
    plt.imshow(Q_table[4][:20,:])
   

    return



```
