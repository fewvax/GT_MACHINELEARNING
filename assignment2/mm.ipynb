```python
def discrete(l, fit, mx, val):
    problem = mlrose_hiive.DiscreteOpt(length = l, fitness_fn = fit, maximize = mx, max_val = val)
    return problem


def mimic(problem, size):
    opt = mlrose_hiive.mimic(problem, pop_size=size, 
                                                 max_attempts=100, max_iters=4000, curve=True, random_state=10)
    return opt
```

```python
def fourpeak(state, l, mx, val, size):
    fitness = mlrose_hiive.FourPeaks(t_pct=0.15)
    problem = discrete(l, fitness, mx, val)
    
    
    
    start = time.time()
    mm = mimic(problem, size)
    mm_t.append(time.time()-start)
    mm_fit.append(mm[1])
```

```python
def contpeak(state, l, mx, val, size):
    fitness = mlrose_hiive.ContinuousPeaks(t_pct=0.15)
    problem = discrete(l, fitness, mx, val)
    
   
    
    start = time.time()
    mm = mimic(problem, size)
    mm_t2.append(time.time()-start)
    mm_fit2.append(mm[1])
```

```python
def ksack(state, l, mx, val, size, weights, values):
    
    
    
    fitness = mlrose_hiive.Knapsack(weights, values, 0.6)
    problem = discrete(l, fitness, mx, val)
    
    

    start = time.time()
    mm = mimic(problem, size)
    mm_t3.append(time.time()-start)
    mm_fit3.append(mm[1])
```

```python
def plotting(labels, data, rang, x, y, title, file):
    plt.figure()
    for label, arr in zip(labels, data):
        plt.plot(rang, arr, label=label)
    plt.legend()
    plt.xlabel(x)
    plt.ylabel(y)
    plt.title(title)
    plt.savefig(file, dpi = 1200)
```

```python
mm_fit =[]

mm_t = []

mm_fit2 =[]

mm_t2 = []

mm_fit3 =[]

mm_t3 = []

for i in [ 40, 80, 160, 320, 640]:
    fourpeak(state, l, True, 2, i)
for i in [ 40, 80, 160, 320, 640]:
    contpeak(state, l, True, 2, i)
for i in [ 40, 80, 160, 320, 640]:
    ksack(state, l, True, 2, i, weights[:l], values[:l])
```

```python
plotting(fit_labels, [mm_fit, mm_fit2, mm_fit3],np.array([ 40, 80, 160, 320, 640]), 
         'Population Size', 'Fitness', 'MIMIC Fitness Value vs. Population Size', 'fp_fit_mmpop.png')
plotting(fit_labels, [mm_t, mm_t2, mm_t3],np.array([ 40, 80, 160, 320, 640]), 
         'Population Size', 'Elapsed Time (s)', 'MIMIC Time vs. Population Size', 'fp_time_mmpop.png')
```
